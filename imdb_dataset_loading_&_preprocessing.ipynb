# Data Cleaning

```python
# Membuat salinan DataFrame untuk proses cleaning yang aman
df_cleaned = df.copy()
```

## Perbaikan data Runtime

dari format '160 min' menjadi '160'.

```python
# Menghapus ' min' dan mengubah Dtype menjadi integer
df_cleaned['Runtime'] = df_cleaned['Runtime'].str.replace(' min', '').astype(int)
```

```python
df_cleaned['Runtime']
```

## Perbaikan data Gross

hilangkan tanda koma, menjadi float

```python
# Menghapus koma dan mengubah Dtype menjadi numerik (float atau integer)
df_cleaned['Gross'] = df_cleaned['Gross'].str.replace(',', '')
df_cleaned['Gross'] = pd.to_numeric(df_cleaned['Gross'], errors='coerce') # errors='coerce' akan mengubah error menjadi NaN

# Setelah itu, kita bisa mengisi NaN yang mungkin muncul dengan 0 seperti rencana awal
df_cleaned['Gross'].fillna(0, inplace=True)
```

```python
df_cleaned['Gross']
```

```python
df_cleaned['Gross'].describe()
```

```python
error_gross_values = df_cleaned[df_cleaned['Gross'].isnull()]
display(error_gross_values)
```

## Perbaikan data Released_Year

menjadi integer

```python
# Mengubah Dtype menjadi numerik, nilai yang salah akan menjadi NaT/NaN
df_cleaned['Released_Year'] = pd.to_numeric(df_cleaned['Released_Year'], errors='coerce')
```

```python
# Diubah menjadi integer saja
df_cleaned['Released_Year'] = pd.to_numeric(df_cleaned['Released_Year'], errors='coerce').fillna(0).astype(int)
```

```python
df_cleaned['Released_Year']
```

```python
zero_released_year_rows = df_cleaned[df_cleaned['Released_Year'] == 0]
display(zero_released_year_rows)
```

Masih ada 1 yang salah input. Perlu diganti jadi 1995.

```python
df_cleaned.loc[df_cleaned['Series_Title'] == 'Apollo 13', 'Released_Year'] = 1995
```

```python
apollo13_row = df_cleaned[df_cleaned['Series_Title'] == 'Apollo 13']
display(apollo13_row)
```

## Perbaikan data Certificates

```python
# Membuat dictionary untuk mapping nilai
cert_mapping = {
    'U/A': 'UA',
    'Approved': 'Passed'
}

# Terapkan mapping ke kolom
df_cleaned['Certificate'] = df_cleaned['Certificate'].replace(cert_mapping)

# Cek kembali hasilnya
print(df_cleaned['Certificate'].value_counts())
```

```python
# Mengisi NaN di kolom Certificate dengan 'Not Rated'
df_cleaned['Certificate'].fillna('Not Rated', inplace=True)
```

```python
df_cleaned['Certificate'].value_counts()
```

```python
# Membuat dictionary untuk mapping nilai
cert_mapping = {
    'Unrated': 'Not Rated'
}

# Terapkan mapping ke kolom
df_cleaned['Certificate'] = df_cleaned['Certificate'].replace(cert_mapping)

# Cek kembali hasilnya
print(df_cleaned['Certificate'].value_counts())
```

## Perbaikan data Meta_score

```python
df_cleaned['Meta_score'].fillna(0, inplace=True)
```
